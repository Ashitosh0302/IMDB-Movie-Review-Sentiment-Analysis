```python
#import some libraries
from tensorflow.keras.preprocessing.text import one_hot
```

```python
#dataset of the sentences
sent=["the glass of milk",
      "the glass of juice",
      "the cup of tea",
      "I am a good boy",
      "I am a good developer",
      "understand the meaning of words",
      "your videos are good",]

sent
```

```python
#DEFINE VOCAB SIZE
voc_size=10000
```

```python
#define the one hot representation of the sentences
one_hot_repre=[one_hot(words,voc_size) for words in sent]
one_hot_repre
```

```python
#WORD EMBEDDING REPRESENTATION
```

```python
#import the library for word embedding
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential

import numpy as np
```

```python
#set the embedding vector features
sent_length=8

embedded_docs=pad_sequences(one_hot_repre,padding="pre",maxlen=sent_length)
print(embedded_docs)
```

```python
#set the dimension
dim=10
```

```python
#now build the model for word embedding
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile("adam","mae")
```

```python
model.summary()
```
